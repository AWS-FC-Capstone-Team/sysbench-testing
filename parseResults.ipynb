```python
import re
import os
import csv
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Number of Threads',
                           'Max Prime',
                           'Events per second',
                           'Total Time',
                           'Total Number of Events',
                           'Min Latency (ms)',
                           'Avg Latency (ms)',
                           'Max Latency (ms)',
                           '95th Percentile Latency (ms)',
                           'Total Execution Time across Threads (s)',
                           'Avg Number of Events per Thread',
                           'Avg Execution Time per Thread (s)'])
```

```python
df
```

```python
for outputFile in os.listdir('./sysbenchOutputs'):
    linenum = 0
    filepath = os.path.join('./sysbenchOutputs', outputFile)
    if outputFile == '.gitkeep':
        continue
    with open(filepath, 'r') as f:
        testResult = []
        for line in f:
            if ':' in line:
                result = line.rsplit(':', 1)[1].strip()
                if 's' in result:
                    testResult.append(float(result[:-1]))
                elif '/' in result:
                    testResult.append(float(result.split('/')[0]))
                elif result != '':
                    testResult.append(float(result))
        df.loc[len(df)] = testResult
```

```python
df = df.astype({'Number of Threads': int, 'Max Prime': int})\
       .sort_values(by=['Number of Threads', 'Max Prime'])
```
